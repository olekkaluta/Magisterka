```python
import cv2
import numpy as np
import os

def calculate_median(points):
    if len(points) == 0:
        return None
    points = np.array(points)
    median_x = int(np.median(points[:, 0]))
    median_y = int(np.median(points[:, 1]))
    return (median_x, median_y)

# Otwieranie pliku wideo
video_path = 'v8.mp4'  # Upewnij się, że podajesz właściwą ścieżkę do swojego pliku wideo
cap = cv2.VideoCapture(video_path)

# Tworzenie folderu do zapisu klatek
output_folder = 'output_frames_blue'
os.makedirs(output_folder, exist_ok=True)

# Plik do zapisu punktów mediany
points_file = open(os.path.join(output_folder, 'median_points.txt'), 'w')

frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Konwersja obrazu do przestrzeni barw HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    # Definiowanie zakresu koloru niebieskiego w przestrzeni HSV
    lower_blue = np.array([100, 150, 50])
    upper_blue = np.array([140, 255, 255])
    
    # Tworzenie maski dla koloru niebieskiego
    mask = cv2.inRange(hsv, lower_blue, upper_blue)
    
    # Znalezienie konturów niebieskich obszarów
    contours, _ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    # Ekstrakcja współrzędnych niebieskich punktów
    blue_points = []
    for contour in contours:
        for point in contour:
            blue_points.append(point[0])
    
    # Obliczenie mediany niebieskich punktów
    median_point = calculate_median(blue_points)
    
    # Zaznaczenie mediany na oryginalnej klatce
    if median_point is not None:
        cv2.circle(frame, median_point, 5, (0, 0, 255), -1)  # Niebieski w BGR to (255, 0, 0)
        # Zapisanie punktu mediany do pliku
        points_file.write(f'{frame_count}: {median_point}\n')
    else:
        # Zapisanie "null" do pliku, jeśli brak punktów
        points_file.write(f'{frame_count}: null\n')
    
    # Zapisanie klatki do folderu
    output_path = os.path.join(output_folder, f'frame_{frame_count:04d}.png')
    cv2.imwrite(output_path, frame)
    
    frame_count += 1
    
    # Wyświetlanie oryginalnej klatki z zaznaczoną medianą
    cv2.imshow('Original Frame with Median', frame)
    
    # Zatrzymanie wideo po naciśnięciu klawisza 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Zwolnienie zasobów
cap.release()
points_file.close()
cv2.destroyAllWindows()
```
