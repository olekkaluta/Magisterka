```python
import numpy as np

# Definiujemy stae
g = 9.81  # Przyspieszenie ziemskie
L1 = 1.0  # Dugo pierwszego wahada
L2 = 1.0  # Dugo drugiego wahada
m1 = 1.0  # Masa pierwszego wahada
m2 = 1.0  # Masa drugiego wahada

def derivatives(state):
    theta1, theta2, omega1, omega2 = state
    
    delta_theta = theta1 - theta2
    
    denominator1 = (2*m1 + m2 - m2 * np.cos(2*delta_theta))
    denominator2 = (L2 / L1) * denominator1
    
    dtheta1 = omega1
    dtheta2 = omega2
    
    domega1 = (-g * (2 * m1 + m2) * np.sin(theta1) - m2 * g * np.sin(theta1 - 2 * theta2) -
               2 * np.sin(delta_theta) * m2 * (omega2**2 * L2 + omega1**2 * L1 * np.cos(delta_theta))) / (L1 * denominator1)
    
    domega2 = (2 * np.sin(delta_theta) * (omega1**2 * L1 * (m1 + m2) + g * (m1 + m2) * np.cos(theta1) +
                                          omega2**2 * L2 * m2 * np.cos(delta_theta))) / (L2 * denominator1)
    
    return np.array([dtheta1, dtheta2, domega1, domega2])

def rk4_step(state, dt):
    k1 = dt * derivatives(state)
    k2 = dt * derivatives(state + 0.5 * k1)
    k3 = dt * derivatives(state + 0.5 * k2)
    k4 = dt * derivatives(state + k3)
    return state + (k1 + 2*k2 + 2*k3 + k4) / 6

# Inicjalizacja stanu pocztkowego
theta1_0 = np.pi / 2  # Pocztkowy kt dla pierwszego wahada
theta2_0 = np.pi / 4  # Pocztkowy kt dla drugiego wahada
omega1_0 = 0.0        # Pocztkowa prdko ktowa dla pierwszego wahada
omega2_0 = 0.0        # Pocztkowa prdko ktowa dla drugiego wahada

state = np.array([theta1_0, theta2_0, omega1_0, omega2_0])

# Definiowanie kroku czasowego
dt = 0.01

# Symulacja dla N krokw
N = 1000
trajectory = np.zeros((N, 4))
trajectory[0, :] = state

for i in range(1, N):
    state = rk4_step(state, dt)
    trajectory[i, :] = state

# Wywietlanie wynikw
import matplotlib.pyplot as plt

time = np.linspace(0, N*dt, N)
theta1_traj = trajectory[:, 0]
theta2_traj = trajectory[:, 1]

plt.plot(time, theta1_traj, label='theta1')
plt.plot(time, theta2_traj, label='theta2')
plt.xlabel('Czas (s)')
plt.ylabel('Kt (rad)')
plt.legend()
plt.show()
```

```python
def make_plot(i, cart_coords, l1, l2, max_trail=30, trail_segments=20, r = 0.05):
    # Plot and save an image of the double pendulum configuration for time step i.
    plt.cla()

    x1, y1, x2, y2 = cart_coords
    ax.plot([0, x1[i], x2[i]], [0, y1[i], y2[i]], lw=2, c='k') # rods
    c0 = Circle((0, 0), r/2, fc='k', zorder=10) # anchor point
    c1 = Circle((x1[i], y1[i]), r, fc='b', ec='b', zorder=10) # mass 1
    c2 = Circle((x2[i], y2[i]), r, fc='r', ec='r', zorder=10) # mass 2
    ax.add_patch(c0)
    ax.add_patch(c1)
    ax.add_patch(c2)

    # plot the pendulum trail (ns = number of segments)
    s = max_trail // trail_segments
    for j in range(trail_segments):
        imin = i - (trail_segments-j)*s
        if imin < 0: continue
        imax = imin + s + 1
        alpha = (j/trail_segments)**2 # fade the trail into alpha
        ax.plot(x2[imin:imax], y2[imin:imax], c='r', solid_capstyle='butt',
                lw=2, alpha=alpha)

    # Center the image on the fixed anchor point. Make axes equal.
    ax.set_xlim(-l1-l2-r, l1+l2+r)
    ax.set_ylim(-l1-l2-r, l1+l2+r)
    ax.set_aspect('equal', adjustable='box')
    plt.axis('off')
    # plt.savefig('./frames/_img{:04d}.png'.format(i//di), dpi=72)

def radial2cartesian(t1, t2, l1, l2):
  # Convert from radial to Cartesian coordinates.
  x1 = l1 * np.sin(t1)
  y1 = -l1 * np.cos(t1)
  x2 = x1 + l2 * np.sin(t2)
  y2 = y1 - l2 * np.cos(t2)
  return x1, y1, x2, y2

def fig2image(fig):
  fig.canvas.draw()
  data = np.fromstring(fig.canvas.tostring_rgb(), dtype=np.uint8, sep='')
  image = data.reshape(fig.canvas.get_width_height()[::-1] + (3,))
  return image
```

```python
L1, L2 = 1, 1
theta1, theta2 = trajectory[:, 0], trajectory[:, 1]
cart_coords = radial2cartesian(theta1, theta2, L1, L2)

fig = plt.figure(figsize=(8.3333, 6.25), dpi=72)
ax = fig.add_subplot(111)

import warnings
warnings.filterwarnings("ignore")

images = []
di = 1
N = 1000
for i in range(0, N, di):
  print("{}/{}".format(i // di, N // di), end='\n' if i//di%20==0 else ' ')
  make_plot(i, cart_coords, L1, L2)
  images.append( fig2image(fig) )
```
